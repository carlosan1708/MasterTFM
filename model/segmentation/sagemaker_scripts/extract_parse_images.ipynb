The following notebook was used to extract the mask from the uncommon folder structure that Sagemaker Ground Truth uses.

```python
import os
import re

dir_name = "SET1_19D"
label_job='Set1-Seg'

manifest_path=dir_name+'/'+label_job+'/'+'manifests/output/output.manifest'

file = open(manifest_path, "r") 
txt=file.readlines()
output_path=dir_name+'/'+label_job+'/'+'annotations/consolidated-annotation/output'
S3_name='tfm-dataset/SET1_19'
im_list=[]
for i in range(len(txt)):

    string = txt[i]
    print(string)
    try:
        im_name = re.search('(?<=SET1_19\/)(.*)(?=.png","Set1-Seg-ref")', string).group(1)
        print('name: ' + im_name)
        im_png=im_name+'.png'
        im_list.append(im_name)
        annotation_name = re.search('output/(.+?)"', string).group(1)
        print('annotation_name: ', annotation_name)
        os.rename("SET1_19/Set1-Seg/annotations/consolidated-annotation/output/" + annotation_name,  "SET1_19/Set1-Seg/annotations/consolidated-annotation/output/"+ im_png)
    except AttributeError:
        print('error')
        pass
```

```python
import numpy as np
from random import sample
  
# Prints list of random items of given length
train_num=len(im_list)*0.8
test_num=len(im_list)*0.2
train_name=sample(im_list,int(train_num))
test_name = list(set(im_list) - set(train_name))
```

```python
print(train_num)
print(test_num)
print(train_name)
print(test_name)
```

```python
os.mkdir('train')
os.mkdir('validation')
os.mkdir('train_annotation')
os.mkdir('validation_annotation')
```

```python
import shutil
for i in range(len(train_name)):
    train_im=train_name[i]+'.png'
    train_im_path=dir_name+'/'+train_im
    train_new_path='train/'+train_im
    print(train_im_path)
    print(train_new_path)
    shutil.move(train_im_path,train_new_path) 
    
    train_annotation=train_name[i]+'.png'
    train_annotation_path=dir_name+'/Set1-Seg/annotations/consolidated-annotation/output/'+train_annotation
    train_annotation_new_path='train_annotation/'+train_annotation
    shutil.move(train_annotation_path,train_annotation_new_path)
```

```python
for i in range(len(test_name)):
    val_im=test_name[i]+'.png'
    val_im_path=dir_name+'/'+val_im
    val_new_path='validation/'+val_im
    print(val_im_path)
    print(val_new_path)
    shutil.move(val_im_path,val_new_path) 
    
    val_annotation=test_name[i]+'.png'
    val_annotation_path=dir_name+'/Set1-Seg/annotations/consolidated-annotation/output/'+val_annotation
    val_annotation_new_path='validation_annotation/'+val_annotation
    print(val_annotation_path)
    print(val_annotation_new_path)
    shutil.move(val_annotation_path,val_annotation_new_path)
```
